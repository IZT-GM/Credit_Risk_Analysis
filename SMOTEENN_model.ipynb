# SMOTEENN Model

```python
import pandas as pd
from path import Path
from collections import Counter
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
data = Path('./Resources/clean_LoanStats.csv')
loans_df = pd.read_csv(data)
loans_df.head()
```

```python
ml_df=loans_df.copy()
ml_df = pd.get_dummies(ml_df, columns=["home_ownership","verification_status","issue_d","pymnt_plan","initial_list_status","next_pymnt_d","application_type","hardship_flag","debt_settlement_flag"])
ml_df.head()
```

```python
# Define features set
X = ml_df.copy()
X = X.drop("loan_status", axis=1)
X.head()
```

```python
# Define target vector
y = ml_df["loan_status"].values
```

```python
# Normal train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
Counter(y_train)
```

## SMOTEENN

```python
# Use the SMOTEENN technique to perform combination sampling on the data
# Count the resampled classes
from imblearn.combine import SMOTEENN

smote_enn = SMOTEENN(random_state=0)
X_resampled, y_resampled = smote_enn.fit_resample(X, y)
Counter(y_resampled)
```

```python
# Fit a Logistic regression model using random undersampled data
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='lbfgs', random_state=1)
model.fit(X_resampled, y_resampled)
```

```python
# Display the confusion matrix
from sklearn.metrics import confusion_matrix

y_pred = model.predict(X_test)
confusion_matrix(y_test, y_pred)
```

```python
# Calculate the Balanced Accuracy Score
from sklearn.metrics import balanced_accuracy_score

balanced_accuracy_score(y_test, y_pred)
```

```python
# Print the imbalanced classification report
from imblearn.metrics import classification_report_imbalanced

print(classification_report_imbalanced(y_test, y_pred))
```
